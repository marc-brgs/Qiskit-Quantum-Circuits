```python
from qiskit import *
```

```python
# Enigme 2 : Les quatres cheveulures
# Lors d’un carnaval, Alice, Bob, Charlie et Dalia décident de participer à un concours pour gagner un voyage inter-spatial dans l'espace. 
# Le défi? Chaque personne doit deviner sa propre couleur de chevelure. 
# L'équipe devra travailler ensemble pour trouver la bonne stratégie afin de déterminer la couleur orange ou indigo de toutes les chevelures.
```

```python
# Initialisation du circuit
q = QuantumRegister(8)
c = ClassicalRegister(1)
circ = QuantumCircuit(q, c)
```

```python
# Définition du circuit
circ.h(0)
circ.h(1)
circ.h(2)
circ.h(3)
circ.barrier()
circ.draw('mpl')
```

```python
# On vérifie si le nombre de couleurs de cheveux indigo devant nous est pair ou impair
circ.cx(1, 4)
circ.cx(2, 4)
circ.cx(3, 4)
circ.barrier()
circ.draw('mpl')
```

```python
# Tout le monde prend note de la réponse
circ.cx(4, 5)
circ.cx(4, 6)
circ.cx(4, 7)
circ.barrier()
circ.draw('mpl')
```

```python
# Bob vérifie la parité de la couleur des cheveux devant lui
circ.cx(2, 5)
circ.cx(3, 5)
circ.barrier()
circ.draw('mpl')
```

```python
# Charlie et Dahlia prennent note de la réponse
circ.cx(5, 6)
circ.cx(5, 7)
circ.barrier()
circ.draw('mpl')
```

```python
# Charlie vérifie la parité de la couleur des cheveux de Dahlia
circ.cx(3, 6)
circ.barrier()
circ.draw('mpl')
```

```python
# Dahlia prend note de la couleur des cheveux de Charlie
circ.cx(6, 7)
circ.draw('mpl')
```
